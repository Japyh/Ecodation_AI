## What is Intelligence?

**The Universal History of Numbers** - Book

* The first machine capable of executing an algorithm without human assistance, the first computer, was developed as a calculator in the 17th century.

* Since then, the complexity of the algorithms that computers can run has gradually increased, and computers have become capable of solving many routine problems in our daily lives. No matter how complex the algorithms they run, it is not possible to say that these computers are intelligent.

* These computers were just automata that repeatedly performed the same simple units over and over again. Intelligence is to be found in how these problem-solving algorithms are generated.

* **Definition 0** -> Intelligence is the ability to correctly and efficiently generate solutions to complex problems in the world we live in. In short, intelligence is the ability to generate algorithms.

## Classical Definition of Artificial Intelligence

* **Definition 1** -> It is a field of computer science concerned with building computers with behaviors that people consider intelligent for each other. This technique has emerged by trying to realize the basic functions of the human brain such as thinking, remembering, evaluating, deciding, comparing and reaching conclusions based on previous experiences in a computer environment.

* **Definition 2** -> It is a systematic science that investigates the properties of intelligent or intelligent systems such as perception, memory, understanding, learning, inference, decision-making and thought generation, suggestion or action for some applications, and aims to build computers or robots with these properties.

* **Definition 3** -> Artificial intelligence is everything that has never been done before.

## The Purpose of Artificial Intelligence

* Providing intelligent systems that help people make decisions and achieve their goals. This is the first level of Artificial Intelligence. `Humanity > Artificial Intelligence`.
    * Perception - Seeing, Listening, Smelling, Touching
    * Reasoning - Thought, Plan, Inquiry
    * Action (Navigation, Maneuvering)
    * Learning (Adaptation, Exploration)
    * Consciousness (Feeling of inwardness)

* Scientific objective: To understand the basics of intelligence and the functions of biological systems.
> For example: How does the brain work

* Engineering objective: To design intelligent machines (programs, autonomous robots...) that can act intelligently in real environments.

* Limits and boundaries: Which jobs are possible? Which tasks are not possible? What is the optimal way to organize a job??

## History of Artificial Intelligence

* In fact, the first time we encounter artificial intelligence is in Ancient Greek mythology with the bronze robot Talos.

<img src="https://i.imgur.com/Y1flYrL.jpg" width=450>

* Talos' mission (as a gift from Zeus) is to tour the shores of the island of Crete three times a day to protect the beautiful Europa, after whom Europe is named.

* Some Chinese and Egyptian legends also describe the existence of such robots.

> For centuries, people have dreamed of building simple robots that can move and do some tasks.
>
> There are sources describing that Al-Jabir, a well-known chemist of the Abbasid period, also built intelligent robots.
>
> There were small robots in the mechanical clocks that Al Jazari gave detailed drawings of in the book he wrote in the early 1200s. These mechanical robots could move their hands and arms and play music.
>
> Legends and later mechanical robots show that humans have always wanted to build machines with artificial intelligence.
>
> In 1637, French mathematician and philosopher Rene Descartes said that a machine that could think like a human would never be built.
>
> The first work recognized as artificial intelligence today is McCulloch and Pitts' Turing-Complete design of artificial neural cells in 1943.
>
> In the 1950s, Shannon and Turing were writing chess programs for computers.
>
> The first neural network-based computer, SNARC, was built by Minsky and Edmonds at MIT in 1951.
>
> John McCarthy, who continued his studies at Princeton University, organized a two-month working meeting at Dartmouth in 1956 with Minsky, Shannon and Rochester.

<br>

> The term `Artificial Intelligence` was first used by `John McCarthy` at the Dartmouth Conference in 1956.
>
> Logic Theorist, one of the first proof-of-theory programs, was introduced here by Newell and Simon.
>
> Later, Newell and Simon developed the `General Problem Solver`, the first program based on the `thinking like a human` approach.
>
> Simon later proposed the physical symbol hypothesis, and this theory became the starting point for those working on building intelligent systems independent of human beings.
>
> The programs written by the researchers participating in the workshop together with their students beat people at checkers, solve algebra problems and speak English.
>
> John McCarthy also developed the Artificial Intelligence programming language Jisp and the concepts of timesharing and e-commerce in computers.
>
> Until the mid-1960s, this work was supported by the US Department of Defense and laboratories were established around the world.
>
> In 1974, with Sir James Lighthill's criticism and the idea of supporting more production-oriented projects, it became difficult to find funding for AI work. The next few years are now known as AI Winter.
>
> On the road to AI Winter, artificial intelligence studies are divided into symbolic and cybernetic AI.
>
> Accordingly, the symbolic AI movement aims to produce machines that think like humans by simulating the syntactic rules of language. However, the AIs produced by this movement fail because they do not take into account the semantic processes of the brain sufficiently.
>
> In the cybernetic artificial intelligence movement, where artificial neural networks are used, the structures used are inadequate because they cannot transform their conclusions into judgments, although they fulfill single-layer tasks.
>
> R1, an artificial intelligence program used by DEC that recommends hardware based on customers' choices, saves the company $40 million a year. This commercial success is helping artificial intelligence efforts to regain momentum.
>
> Lisp became the language most preferred by those who write artificial intelligence programs. The Lisp language has continuously evolved until today and dialects such as ANSI Common Lisp, Emacs Lisp, Scheme and Autolisp have been written. Thanks to the Lisp programming language, voice recognition technology was developed and the personal assistant application of the Iphone was created (today).
>
> Inspired by Japan's fifth generation computer project, the US and the UK increased their support again, and the artificial intelligence industry, which reached 1 billion dollars in 1985, exceeded 2 billion dollars in 1988. However, with the collapse of the Lisp machine market in 1987, AI research entered a second and longer pause.
>
> In the late 1990s and early millennium, artificial intelligence was being applied to fields as diverse as logistics, data mining and medical diagnosis.
>
> On May 11, 1997, Deep Blue became the first chess-playing computer to defeat Garry Kasparov, a world chess champion.

## Why Do We Need Data Science?

* It is estimated that 90 percent of the world's data has been created in the last two years. For example, Facebook users upload 10 million photos every hour.

* The wealth of stored data can offer transformational advantages to organizations and societies around the world, but only if it can be interpreted.

* Rather than relying on business analysts to see what they can discover from the data, machine learning (ML) models are now much more preferred to learn from large amounts of uploaded data.

* Data is the cornerstone of innovation, but its value comes from the insights that data scientists can gain from it and the actions that can be taken based on it.

## How and where do we start?

Learning SQL is an excellent way to get started in Data Science. Not only is it the heart and soul of every Data Scientist, but it also teaches you:

* Teaches to think in terms of relational tables

* You understand how a database works

* Translate business problems into accurate data sources

* You learn to connect different pieces of information.

If you want to learn more about data, start with SQL. Because you won't just learn a programming language. You will also learn how to think differently about problem solving!

<img src="https://upload.wikimedia.org/wikipedia/commons/8/87/Sql_data_base_with_logo.png" width=300>

<br>

There are other practical benefits to learning SQL for Data Science before Excel. Basically, what we do in both is the same:

* Organizing Data

* Collecting Data

* Combining Data

* Grouping/Sorting Data

But with SQL it is possible to do this MUCH faster and at a larger scale. Therefore, it is MUCH better to first learn how to properly process your data with SQL... Then you can apply Excel when needed. This will save you a lot of frustration and make you a better data analyst all round!

## NumPy

<img src="https://upload.wikimedia.org/wikipedia/commons/thumb/1/1a/NumPy_logo.svg/2560px-NumPy_logo.svg.png" width=300>

* It provides objects for multidimensional arrays and matrices, as well as functions to easily perform advanced mathematical and statistical operations on these objects.

* It enables vectorization of mathematical operations on arrays and matrices, which significantly improves performance.

* Many other Python libraries are built on NumPy.

* It is indispensable for high-performance vector and matrix calculations.

* Contains precompiled functions for numerical calculation routines.

* Written in C and Fortran. C and Fortran code can also be embedded.

**Note**: The more user-friendly a programming language is and the more colloquial its syntax is, the slower it is.

## Exploratory Data Analysis (EDA)

* **EDA** is a crucial component of a data science project.

* Although it is a crucial step, it is often overlooked or underestimated by many practitioners.

* John Tukey provided guidance on statistical analysis of data sets in his book Exploratory Data Analysis.

* EDA is also used to prepare your analysis for advanced modeling once you have defined the basic data characteristics and know what questions to ask about your data.

* EDA uses quantitative methods to help you visualize your data to identify patterns, anomalies and data characteristics and draw conclusions.
